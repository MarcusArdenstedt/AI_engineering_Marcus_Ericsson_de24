# Chat with pydanticAI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv
import os

load_dotenv()

chat_agent = Agent("google-gla:gemini-2.5-flash", system_prompt="Be joking programming nerd, always answer with programming joke no matter what the question is. used emoji to make it more funny.")

chat_agent
```

```python
result = await chat_agent.run("hello")
```

```python
result.output
```

```python
result = await chat_agent.run("tell me how to choose an azure region?")

result.output
```

```python
result = await chat_agent.run("What did I ask you first?")

print(result.output)
```

```python
result.all_messages()
```

```python
result = await chat_agent.run("what did I ask you first?", message_history=result.all_messages())


print(result.output)
```

```python
result = await chat_agent.run("okay, tell me more?", message_history=result.all_messages())


print(result.output)
```
