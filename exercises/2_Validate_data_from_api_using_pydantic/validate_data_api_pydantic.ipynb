# Validate data from API using Pydan

a) Create a Pydantic model with name Joke with the following fields

- id with type integer
- joke with type string

```python
import requests

headers = {"Accept": "application/json"}
response = requests.get("https://icanhazdadjoke.com/", headers=headers)

print(response.json())
```

```python

data = response.json()

data
    
    
```

```python
data["id"], data["joke"]
```

```python
from pydantic import BaseModel, ValidationError

class Joke(BaseModel):
    id: int
    joke: str
    
    
```

```python

try:
    joke1 = Joke.model_validate(data)
except ValidationError as err:
    print(err)

```

```python
from pydantic import BaseModel, computed_field

class Joke(BaseModel):
    id: int
    joke: str
    
    @computed_field
    @property
    def words_in_joke(self) -> int:
        """return numner of words in the joke"""
        return len(self.joke.split())
```

```python
joke2 = Joke.model_validate(data)

joke2
```
