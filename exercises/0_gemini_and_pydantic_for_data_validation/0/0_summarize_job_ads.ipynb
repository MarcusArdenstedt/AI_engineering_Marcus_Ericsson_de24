```python
from pathlib import Path


directory = Path.cwd().parents[1] / "data"

ads = {}
for ad in range(1, 3 + 1):
    data_job_ads = directory/f"ads{ad}.txt"
    with open(data_job_ads, "r", encoding="UTF-8") as file:
        data = file.read()
        ads[ad]= data
     

ads
```

```python
from google.genai import types




def summarize_job_ads(data, model= "gemini-2.5-flash"):
    response = client.models.generate_content(
    model=model,
    contents="Summarize the text shorly, write it in markdown format",
    config=types.GenerateContentConfig(
        system_instruction=data
        
    )
)

    return response.text 


Path("summarize").mkdir(exist_ok=True)
for ad in range(1, 4):
    test = summarize_job_ads(data= ads[ad])

    with open(f"summarize/ads{ad}.md", "w") as file:
        data = file.write(test)

```

```python

from pathlib import Path

def summarize_text(ad):
    my_file = client.files.upload(file= ad)

    response = client.models.generate_content(
        model= "gemini-2.5-flash",
        contents=["Summarize the text shortly, type out in markdown format", my_file]
    )
    return response.text

Path("summarize_with_gemini").mkdir(exist_ok=True)
for ad_nr in range(1, 4):
    directory = Path.cwd().parents[1] / "data"
    dir_ads = directory / f"ads{ad_nr}.txt"
    
    data = summarize_text(ad = dir_ads)
    
    with open(f"summarize_with_gemini/ad{ad_nr}.md", "w") as file:
        file.write(data)
    
    
    
```
