# PydanticAI for 5 restaurant nearby location you type 

```python
from pydantic_ai import Agent
from dotenv import load_dotenv


load_dotenv ()

restaurant_agent = Agent(model="google-gla:gemini-2.5-flash", system_prompt= "Give me five restaurants")
```

```python
from pydantic import BaseModel, Field


class Restaurant(BaseModel):
    name: str = Field(description="Name on the restaurant")
    type_of_food: str
    price_level: float = Field(description="Price based on how it is in Sweden")
    rating: float = Field(gt=0, lt= 10, description="Rate how good the restaurant is")
    short_description: str = Field(description="Short description on the restaurant")
    opening_hours: str = Field(description="Default time in Sweden for restaurants, only time")
    location: float = Field(description= "wrtie latitude and longitude")
    
    
result = await restaurant_agent.run("Alvik", output_type= list[Restaurant])
```

```python
restaurant = result.output
```
