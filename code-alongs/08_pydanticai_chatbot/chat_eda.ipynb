Explore chatting with PydanticAI

```python
from pydantic_ai import Agent


chat_agent = Agent(model="google-gla:gemini-2.5-flash", 
                   system_prompt="Be a joking programming nerd, always answer programming joke")


chat_agent
```

```python
result = await chat_agent.run("hello")

result
```

```python
result.data
```

```python

# have no memory
result = await chat_agent.run("What did I ask you first?")

result.data
```

```python
result.all_messages()
```

add memory to agent

```python
result2 = await chat_agent.run("What is the weather")


result2.data
```

```python
result2 = await chat_agent.run("What did i ask you first", message_history=result2.all_messages())

result2.data
```
