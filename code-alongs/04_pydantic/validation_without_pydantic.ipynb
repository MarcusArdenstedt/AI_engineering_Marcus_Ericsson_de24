# Validation

```python
# Student class
# name, age, gender


# instantiate instance from Student

class Student:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
        
    @property
    def age(self):
        return self._age 
    
    
    @age.setter
    def age(self, age):
        # Validation code here
        
        if not isinstance(age, (int, float)):
            raise TypeError(f"age must be number not {type(age)}")
        
        if not 0 <= age < 125:
            raise ValueError(f"Age must be between 0 and 124, not {type(age)}")
         
        self._age = age
    
        
        
    # "dunder repper" -> representation of this instance 
    def __repr__(self):
        return f"Student(name= '{self.name}', age= {self.age}, gender= '{self.gender}')"
        
        
try:  
    Student(name="Adam", age= "femtioa책tta", gender="M")
    
except TypeError as err:
    print(err)
 


        
```

```python
# Student class
# name, age, gender


# instantiate instance from Student

class Student:
    def __init__(self, name, age, gender):
        self.name = name  # attribute
        self.age = age  # porperty
        self.gender = gender  # attribute
        
    @property
    def age(self):
        return self._age 
    
    
    @age.setter
    def age(self, age):
        # Validation code here
        
        if not isinstance(age, (int, float)):
            raise TypeError(f"age must be number not {type(age)}")
        
        if not 0 <= age < 125:
            raise ValueError(f"Age must be between 0 and 124, not {type(age)}")
         
        self._age = age
    
    # special method
    # "dunder repper" -> representation of this instance 
    def __repr__(self):
        return f"Student(name= '{self.name}', age= {self.age}, gender= '{self.gender}')"
        
        
try:  
    Student(name="Adam", age= "femtioa책tta", gender="M")
    
except TypeError as err:
    print(err)
 
s3=  Student(name="Adam", age= 124, gender="M")
```
